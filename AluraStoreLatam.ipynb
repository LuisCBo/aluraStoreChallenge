### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
total_tienda = []
noTienda = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
total_tienda.append(tienda.Precio.sum())
total_tienda.append(tienda2.Precio.sum())
total_tienda.append(tienda3.Precio.sum())
total_tienda.append(tienda4.Precio.sum())

import matplotlib.pyplot as plt

plt.plot(noTienda, total_tienda, 'x', color='red')
plt.title("Total facturado por tienda")
plt.xlabel("Tienda")
plt.ylabel("Total")
plt.show()
```

```python
print("Total tienda 1: " + str(total_tienda[0]))
print("Total tienda 2: " + str(total_tienda[1]))
print("Total tienda 3: " + str(total_tienda[2]))
print("Total tienda 4: " + str(total_tienda[3]))
```

# 2. Ventas por categoría

```python
ventas_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_tienda4 = tienda4['Categoría del Producto'].value_counts()

plt.figure(figsize=(15, 6))
plt.subplot(1, 3, 1)
plt.plot(ventas_tienda1)
plt.plot(ventas_tienda1, 'x', color='red')
plt.title("Ventas por categoría de la tienda 1")
plt.xticks(rotation=90)
plt.xlabel("Categoría")
plt.ylabel("Ventas")
plt.ylim(100, 500)

plt.subplot(1, 3, 2)
plt.plot(ventas_tienda2)
plt.plot(ventas_tienda2, 'x', color='red')
plt.title("Ventas por categoría de la tienda 2")
plt.xticks(rotation=90)
plt.xlabel("Categoría")
plt.ylabel("Ventas")
plt.ylim(100, 500)

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(1, 3, 1)
plt.plot(ventas_tienda3)
plt.plot(ventas_tienda3, 'x', color='red')
plt.title("Ventas por categoría de la tienda 3")
plt.xticks(rotation=90)
plt.xlabel("Categoría")
plt.ylabel("Ventas")
plt.ylim(100, 500)

plt.subplot(1, 3, 2)
plt.plot(ventas_tienda4)
plt.plot(ventas_tienda4, 'x', color='red')
plt.title("Ventas por categoría de la tienda 4")
plt.xticks(rotation=90)
plt.xlabel("Categoría")
plt.ylabel("Ventas")
plt.ylim(100, 500)

plt.tight_layout()
plt.show()
```

```python
print("Ventas tienda 1: ")
print(ventas_tienda1)
print("Ventas tienda 2: ")
print(ventas_tienda2)
print("Ventas tienda 3: ")
print(ventas_tienda3)
print("Ventas tienda 4: ")
print(ventas_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = []
calificacion_tienda.append(round(tienda.Calificación.mean(), 2))
calificacion_tienda.append(round(tienda2.Calificación.mean(), 2))
calificacion_tienda.append(round(tienda3.Calificación.mean(), 2))
calificacion_tienda.append(round(tienda4.Calificación.mean(), 2))

plt.plot(noTienda, calificacion_tienda, 'x', color='red')
plt.title("Calificación promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación")
plt.ylim(3.9, 4.1)
plt.show()
```

```python
print("Calificación tienda 1: " + str(calificacion_tienda[0]))
print("Calificación tienda 2: " + str(calificacion_tienda[1]))
print("Calificación tienda 3: " + str(calificacion_tienda[2]))
print("Calificación tienda 4: " + str(calificacion_tienda[3]))
```
